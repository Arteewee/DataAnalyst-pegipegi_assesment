# 1.2	Find the average time (in minute) it takes users to make payment of their bookings. Present the average times for each payment method.

```
df['time_booking_created'] = pd.to_datetime(df['booking_created_time']).dt.time
```

```
df['time_booking_paid'] = pd.to_datetime(df['booking_paid_time']).dt.time
```

```
df['booking_created_time_minute'] = df['time_booking_created'].astype(str).str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]))
```

```
df['booking_paid_time_minute'] = df['time_booking_paid'].astype(str).str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]))
```

```
df['time_user_pay'] = df['booking_paid_time_minute'] - df['booking_created_time_minute']
```

```
df
```

```
avg_time = df[['payment_method', 'time_user_pay']]
```

```
avg_time
```

```
df2 = avg_time.groupby(['payment_method']).mean('time_user_pay')
```

```
df2
```

# 1.3	Get the first five flight ticket purchases of each user in 2017 who have purchased flight tickets at least 5 times and have spent IDR 25 Mio. Sort from users who have highest spend, and then sort by the purchase sequence (from 1 to 5) for each user.

```
df['booking_created_date'] = pd.to_datetime(df['booking_created_time']).dt.date
```

```
df['total_booking_price_amount'] = df.groupby('user_id').booking_price_amount.transform(np.sum)
```

```
df['booking_sequence'] = df.sort_values(['user_id'], ascending=[False]) \
             .groupby(['user_id']) \
             .cumcount() + 1
```

```
df
```

```
flight_ticket = df[['user_id', 'booking_sequence', 'booking_created_date', 'booking_price_amount', 'total_booking_price_amount']]
```

```
flight_ticket
```

```
df3 = flight_ticket[flight_ticket['booking_sequence']<=5][flight_ticket['total_booking_price_amount']>25000000][(flight_ticket['booking_created_date'] >= pd.to_datetime('2017-01-01').date()) & (flight_ticket['booking_created_date'] <= pd.to_datetime('2017-12-31').date())].sort_values(['user_id','booking_sequence'],ascending=[False, True])
```
